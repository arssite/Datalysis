Spam classification using the re corpus and nltk library involves the following steps:
Data Preparation:

Import the re corpus and nltk library. Load the reuter corpus using nltk.corpus.reuters.fileids(). Filter the file IDs to only include documents from the training and test sets. Use nltk.corpus.reuters.raw(fileid) to access the raw text of each document. Feature Extraction:

Use regular expressions to extract features from the text. The re corpus provides a categories() function to access the categories of documents. Use this information to label the data for training. Model Training:

Use the extracted features and labels to train a spam classifier. nltk provides various machine learning algorithms for classification, such as NaiveBayesClassifier. Evaluation:

Use the trained classifier to predict the category of new documents. Calculate the accuracy of the classifier by comparing its predictions with the actual labels.



```python
import pandas as pd
data=pd.read_csv('https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv', encoding='latin-1')
```

```python
#droping NaN Columns
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
data.columns=['label','text']
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
import nltk
nltk.download('all')
```

```python
txt=list(data['text'])
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(txt)):
  r=re.sub('[^a-zA-Z]',' ',txt[i])
  r=r.lower()
  r=r.split()
  r=[word for word in r if word not in stopwords.words('english')]
  r=[lem.lemmatize(word) for word in r]
  r=' '.join(r)
  corpus.append(r)
```

```python
data['txt']=corpus
data.head()
```

```python
#feature & label Set
x=data['text']
y=data['label']
```

```python
#spliting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=123)
```

```python
print("Train",x_train.shape)
print("Test",x_test.shape)
```

# Bag of Words
(BoW) is a method used in Natural Language Processing (NLP) to represent text data as a collection of individual words, disregarding grammar and word order. It is a simple but effective way to convert textual data into a numerical format that can be used by machine learning algorithms.

Steps involved in Bag of Words:

Preprocessing:

Clean the text data by removing punctuation, stop words, and converting all words to lowercase. Tokenization:

Split the preprocessed text into individual words or tokens. Dictionary Creation:

Create a dictionary of all unique words in the corpus. Vectorization:

For each document, create a vector of the same length as the dictionary. Each element in the vector represents the frequency of the corresponding word in the document. Example:

Consider the following two sentences:

"The quick brown fox jumps over the lazy dog." "The dog barked at the fox." Preprocessing:

Remove punctuation and stop words: "quick brown fox jumps lazy dog", "dog barked fox". Tokenization:

Split into tokens: ["quick", "brown", "fox", "jumps", "lazy", "dog"], ["dog", "barked", "fox"]. Dictionary Creation:

Unique words: {"quick", "brown", "fox", "jumps", "lazy", "dog", "barked"}. Vectorization:

Sentence 1: [1, 1, 1, 1, 1, 2] Sentence 2: [1, 0, 1, 0, 0, 1] Advantages of Bag of Words:

Simple and easy to implement. Captures the presence or absence of words in a document. Can be used with various machine learning algorithms. Disadvantages of Bag of Words:

Ignores word order and grammar. Doesn't capture the context of words. Can lead to high dimensionality if the vocabulary is large.

```python
#training logistic Regression Model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_train_cv=cv.fit_transform(x_train)
x_train_cv.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train_cv,y_train)

x_test_cv=cv.transform(x_test)
```

```python
#generate predictions
pred=lr.predict(x_test_cv)
pred
```
