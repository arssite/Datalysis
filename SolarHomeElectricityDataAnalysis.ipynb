
Q1.**How has energy consumption evolved over time in each consumption category?**


```python
df['date'] = pd.to_datetime(df['date'])
```

```python
category_consumption = df.groupby(['Consumption Category', df['date'].dt.year])['Generator Capacity'].sum().unstack(level=0)
```

```python
category_consumption.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.title('Energy Consumption by Category Over Time')
plt.legend()
plt.show()
```

**Q2. Are there any noticeable seasonal patterns or trends in energy consumption within different consumption categories?**

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
category_consumption = df.groupby(['Consumption Category', 'year', 'month'])['Generator Capacity'].mean().unstack(level=0)
```

```python
category_consumption.plot(subplots=True, figsize=(10, 10))
plt.xlabel('Month')
plt.ylabel('Energy Consumption')
plt.suptitle('Seasonal Patterns of Energy Consumption by Category')
plt.show()
```

Q3. **Predict future energy consumption trend within each consumption category?**

```python
category_consumption = df.groupby(['Consumption Category', df['date'].dt.date])['Consumption Category'].sum().unstack(level=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
category = 'CL'  # Replace 'Your_Category' with the consumption category you want to forecast
df_category = category_consumption[[category]].copy()
df_category.set_index(pd.to_datetime(df_category.index), inplace=True)
df_category = df_category.resample('M').sum()  # Resample to monthly frequency

# Step 5: Split the data into training and testing sets
train_size = int(len(df_category) * 0.8)
train_data, test_data = df_category[:train_size], df_category[train_size:]

# Step 6: Train an ARIMA model on the training data
model = ARIMA(train_data, order=(2, 1, 2))
model_fit = model.fit()

# Step 7: Make predictions on the testing data and evaluate the model's performance
predictions = model_fit.forecast(steps=len(test_data))[0]
mae = mean_absolute_error(test_data, predictions)
rmse = np.sqrt(mean_squared_error(test_data, predictions))
print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}")

# Step 8: Visualize the predicted and actual energy consumption trends
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Actual Data')
plt.plot(test_data.index, predictions, label='Predictions')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.title(f'{category} Energy Consumption Forecast')
plt.legend()
plt.show()
```
