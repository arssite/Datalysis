# Spam classification using the re corpus and nltk library involves the following steps:

Data Preparation:

Import the re corpus and nltk library.
Load the reuter corpus using nltk.corpus.reuters.fileids().
Filter the file IDs to only include documents from the training and test sets.
Use nltk.corpus.reuters.raw(fileid) to access the raw text of each document.
Feature Extraction:

Use regular expressions to extract features from the text.
The re corpus provides a categories() function to access the categories of documents.
Use this information to label the data for training.
Model Training:

Use the extracted features and labels to train a spam classifier.
nltk provides various machine learning algorithms for classification, such as NaiveBayesClassifier.
Evaluation:

Use the trained classifier to predict the category of new documents.
Calculate the accuracy of the classifier by comparing its predictions with the actual labels.

```python
import pandas as pd
data=pd.read_csv('https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv', encoding='latin-1')
```

```python
#droping NaN Columns
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
data.columns=['label','text']
data.head()
```

```python
import nltk
nltk.download('all')
```

```python
txt=list(data['text'])
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(txt)):
  r=re.sub('[^a-zA-Z]',' ',txt[i])
  r=r.lower()
  r=r.split()
  r=[word for word in r if word not in stopwords.words('english')]
  r=[lem.lemmatize(word) for word in r]
  r=' '.join(r)
  corpus.append(r)
```

```python
data['txt']=corpus
data.head()
```
