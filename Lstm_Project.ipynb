# LSTM stands for Long Short-Term Memory,
a type of recurrent neural network (RNN) that is well-suited for processing sequential data. LSTMs are designed to address the vanishing gradient problem, which can occur in traditional RNNs when dealing with long sequences.

Here's how LSTMs work:

Cell state: LSTMs have a cell state that acts as a conveyor belt, carrying information through the network. The cell state is updated at each time step based on the current input and the previous cell state.
Input gate: The input gate controls the flow of information into the cell state. It decides how much of the current input and the previous cell state should be added to the current cell state.
Forget gate: The forget gate controls how much of the previous cell state should be forgotten. This allows the LSTM to learn long-term dependencies by selectively forgetting irrelevant information.
Output gate: The output gate controls the flow of information from the cell state to the output of the LSTM. It decides how much of the current cell state should be exposed to the subsequent layers in the network.
LSTMs are commonly used in a variety of applications, including:

Natural language processing (NLP): LSTMs are used in NLP tasks such as machine translation, text classification, and sentiment analysis.
Speech recognition: LSTMs are used in speech recognition systems to recognize spoken words.
Time series forecasting: LSTMs are used in time series forecasting to predict future values based on past data.

*Next word prediction using LSTM works by training the LSTM model on a large corpus of text data. The model learns the relationships between words and sequences of words, and uses this knowledge to predict the next word in a sequence.*


____________________________________________________________

Here's how next word prediction using LSTM works:

**Tokenize the text data:** The first step is to tokenize the text data into individual words or subwords.

**Create training data:** The training data consists of pairs of input sequences and target words. For example, if the input sequence is "The cat sat on the", the target word could be "mat".

**Train the LSTM model:** The LSTM model is trained on the training data using a technique called backpropagation through time. This allows the model to learn the long-term dependencies between words in the sequence.

**Make predictions:** Once the model is trained, it can be used to predict the next word in a sequence. To do this, the model is given an input sequence and then generates a probability distribution over all possible next words. The word with the highest probability is then predicted as the next word.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len  = max([len(x) for x in input_sequences])
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
X
```

```python
y = padded_input_sequences[:,-1]
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=283)
```

```python
y.shape
```

```python
model = Sequential()
model.add(Embedding(283, 100, input_length=56))
model.add(LSTM(150))
model.add(Dense(283, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

```python
model.fit(X,y,epochs=100)
```

```python
import numpy as np
import time
```

```python
txt="What is the fee"
for i in range(10):
  tokentxt=tokenizer.texts_to_sequences([txt])[0]
  padtokentxt=pad_sequences([tokentxt],maxlen=56,padding='pre')
  pos=np.argmax(model.predict(padtokentxt))

  for word,index in tokenizer.word_index.items():
    if index==pos:
      txt=txt+" "+word
      print(txt)
      time.sleep(2)
```

```python
def nextword(txt):
  for i in range(10):
    tokentxt=tokenizer.texts_to_sequences([txt])[0]
    padtokentxt=pad_sequences([tokentxt],maxlen=56,padding='pre')
    pos=np.argmax(model.predict(padtokentxt))

    for word,index in tokenizer.word_index.items():
      if index==pos:
        txt=txt+" "+word
        print(txt)
        time.sleep(1)
```
